```python
import requests
import pandas as pd
import yfinance as yf
# Lista de tickers dos ativos
tickers = pd.read_csv("tickers.csv")
tickers = tickers['Código'].to_list()
# Baixar dados históricos para os ativos
start_date = '2017-01-01'
end_date = '2023-11-19'
data = yf.download(tickers, start=start_date, end=end_date)['Adj Close']
data['time'] = data.index
data
```

```python
# Use a função melt para transformar o DataFrame
data = pd.melt(data, id_vars=['time'], var_name='Ativo', value_name='Preço')
data
```

```python
data['retorno_12m'] = data.groupby('Ativo')['Preço'].transform(lambda x: x / x.shift(252) - 1)
data['Lagged_Preço'] = data.groupby('Ativo')['Preço'].shift()
data
```

```python

# Converta a coluna 'time' para o tipo de datetime se ainda não for
data['time'] = pd.to_datetime(data['time'])

# Use a função groupby para agrupar por 'Ativo' e o ano-mês de 'time'
groups = data.groupby(['Ativo', data['time'].dt.to_period('M')])

# Use a função idxmax para obter o índice do máximo para cada grupo
idx = groups['time'].idxmax()

# Crie o DataFrame 'data_mensal' usando os índices encontrados
data_mensal = data.loc[idx]
data_mensal
```

```python
# Crie a coluna 'Yearmon'
data_mensal['Yearmon'] = data_mensal['time'].dt.to_period('M')
data_mensal
```

```python
data_mensal['rank'] = data_mensal.groupby('Yearmon')['retorno_12m'].rank(ascending=False)
data_mensal
```

```python
# Filtrar os 10 maiores de cada grupo 'Yearmon'
top_10_retornos = data_mensal[data_mensal['rank'].between(1, 10)].reset_index(drop=True)
top_10_retornos
```

```python
top_10_retornos['time1'] = pd.to_datetime(top_10_retornos['time']) + pd.DateOffset(months=1)
top_10_retornos['Yearmon_c'] = top_10_retornos['time1'].dt.to_period('M')
top_10_retornos
```

```python
top_10_retornos = top_10_retornos[['Yearmon_c', 'rank', 'Ativo']]
top_10_retornos
```

```python
data['time'] = pd.to_datetime(data['time'])
data['Yearmon_c'] = data['time'].dt.to_period('M')
data
```

```python
data = pd.merge(data, top_10_retornos[['Yearmon_c', 'rank', 'Ativo']], how='inner', on=['Yearmon_c', 'Ativo'])
data
```

```python
data['retorno_grupo'] = data.groupby('rank').apply(lambda group: (group['Preço'] / group['Lagged_Preço']) - 1).reset_index(level=0, drop=True)
data['cota'] = data.groupby(['rank', 'Yearmon_c'])['retorno_grupo'].transform(lambda x: (1 + x).cumprod())
data
```

```python
data = (
    data.groupby('time')
    .agg(cota_long=('cota', lambda x: x.sum() / len(x)))
    .reset_index()
)
data
```

```python
import requests
import pandas as pd
import yfinance as yf

# Example endpoint for fetching cryptocurrency data
url = "https://api.coingecko.com/api/v3/coins/markets"

# Example parameters (adjust as needed)
params = {
    'vs_currency': 'usd',
    'order': 'market_cap_desc',
    'per_page': 100,
    'page': 1,
    'sparkline': False,
}

# Make the API request
response = requests.get(url, params=params)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the response JSON
    data = response.json()

    # Extract the 'Symbol' column as a list
    symbols_list = [entry['symbol'] + '-USD' for entry in data]

    # Create a DataFrame with the 'Symbol' list
    df = pd.DataFrame({'Symbol': symbols_list})

    # Print the DataFrame
    print(df)

    # Extract the symbols as a list
    symbols_to_download = df['Symbol'].tolist()

    # Download historical data using yfinance
    start_date = '2017-01-01'
    end_date = '2023-11-19'
    data = yf.download(symbols_to_download, start=start_date, end=end_date)['Adj Close']

    # Print the historical data
    print(data)
else:
    print("Failed to fetch data. Status code:", response.status_code)
```

```python
data['time'] = data.index
# Use a função melt para transformar o DataFrame
data = pd.melt(data, id_vars=['time'], var_name='Ativo', value_name='Preço')
data['retorno_12m'] = data.groupby('Ativo')['Preço'].transform(lambda x: x / x.shift(252) - 1)
data['Lagged_Preço'] = data.groupby('Ativo')['Preço'].shift()

# Converta a coluna 'time' para o tipo de datetime se ainda não for
data['time'] = pd.to_datetime(data['time'])

# Use a função groupby para agrupar por 'Ativo' e o ano-mês de 'time'
groups = data.groupby(['Ativo', data['time'].dt.to_period('M')])

# Use a função idxmax para obter o índice do máximo para cada grupo
idx = groups['time'].idxmax()

# Crie o DataFrame 'data_mensal' usando os índices encontrados
data_mensal = data.loc[idx]
data_mensal['time'] = pd.to_datetime(data_mensal['time'])

# Crie a coluna 'Yearmon'
data_mensal['Yearmon'] = data_mensal['time'].dt.to_period('M')
data_mensal['rank'] = data_mensal.groupby('Yearmon')['retorno_12m'].rank(ascending=False)

# Filtrar os 10 maiores de cada grupo 'Yearmon'
top_10_retornos = data_mensal[data_mensal['rank'].between(1, 10)].reset_index(drop=True)
top_10_retornos
```

```python
top_10_retornos['time1'] = pd.to_datetime(top_10_retornos['time']) + pd.DateOffset(months=1)
top_10_retornos['Yearmon_c'] = top_10_retornos['time1'].dt.to_period('M')
top_10_retornos = top_10_retornos[['Yearmon_c', 'rank', 'Ativo']]
top_10_retornos
data['time'] = pd.to_datetime(data['time'])
data['Yearmon_c'] = data['time'].dt.to_period('M')
data = pd.merge(data, top_10_retornos[['Yearmon_c', 'rank', 'Ativo']], how='inner', on=['Yearmon_c', 'Ativo'])
data['retorno_grupo'] = data.groupby('rank').apply(lambda group: (group['Preço'] / group['Lagged_Preço']) - 1).reset_index(level=0, drop=True)
data['cota'] = data.groupby(['rank', 'Yearmon_c'])['retorno_grupo'].transform(lambda x: (1 + x).cumprod())
data = (
    data.groupby('time')
    .agg(cota_long=('cota', lambda x: x.sum() / len(x)))
    .reset_index()
)


# Criando a coluna 'Yearmon'
data['Yearmon'] = data['time'].dt.to_period('M')

# Adicionando a coluna 'order'
data['order'] = data.groupby('Yearmon')['time'].rank()
# Calculando 'return_long'
data['return_long'] = data.apply(lambda row: row['cota_long'] - 1.0 if row['order'] == 1 else row['cota_long'] / data.loc[row.name - 1, 'cota_long'] - 1, axis=1)
data['cota_portfolio'] = (1 + data['return_long']).cumprod()
data
```
