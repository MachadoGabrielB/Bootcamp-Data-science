```python
#tratando coluna 'role'

def tratamento_coluna (data: pd.DataFrame, serie:str, get:str) :
    
    series = list(data[serie])
    new_list = list()
    for elem in series :
        new_list.append(eval(elem).get(get))
    data[serie] = new_list

    return data
```

```python
data = tratamento_coluna (data=data, serie='role', get='S')
```
