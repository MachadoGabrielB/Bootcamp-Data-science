```python
#tratando coluna 'role'

def tratamento_coluna (data: pd.DataFrame, serie:str, get:str) :
    
    series = list(data[serie])
    new_list = list()
    for elem in series :
        new_list.append(eval(elem).get(get))
    data[serie] = new_list

    return data
```

```python
data = tratamento_coluna (data=data, serie='role', get='S')
```

```python
series = data["natural_percentage"].apply( lambda x: eval(x))
```

```python
dic = {
    "F" : [],
    "L" : [],
    "O" : [],
    "W" : []
}
for line in list(series) : 
    list_ = line["S"]
    dict_ = eval(list_)[0]

    for key, values in dict_.items() :
        dic[key].append(values)
    
for key, value in dic.items() :
    data[f"natural_{key.lower()}"] = value
```
