# Webscrapping

```python
import requests as req
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://portalnovonorte.com.br/educacao'
pagina = req.get(url)
```

```python
pagina.text
```

```python
soup = bs(pagina.text)
soup
```

```python
conteudo = soup.find(class_= 'conteudo-interno')
conteudo
```

```python
lista_divs = conteudo.find_all('a', href = True)
lista_divs
```

```python
lista_divs[0]['href']
```

```python
lista_link = []
for elemento in lista_divs:
    lista_link.append(elemento['href'])
```

```python
lista_link
```

```python
from newspaper import Article
```

```python
url1 = lista_link[0]
article = Article(url1)
```

```python
article.download()
article.parse()
```
